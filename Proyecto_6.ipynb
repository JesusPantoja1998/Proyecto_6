# Prepara los datos

```python
games_data.columns=games_data.columns.str.lower() #se convierte el nombre de todas las columnas a minusculas
games_data["year_of_release"] = pd.to_datetime(games_data["year_of_release"], format='%Y')# se convierte la columna year_of_release a tipo fecha
#games_data['year_of_release'] = games_data['year_of_release'].dt.strftime('%Y')
#games_data['year_of_release'] = games_data['year_of_release'].dt.year
games_data["user_score"] = pd.to_numeric(games_data["user_score"], errors='coerce')
games_data["user_score"] = games_data["user_score"].fillna(0)
games_data["critic_score"] = games_data["critic_score"].fillna(0)
sales_columns = ['na_sales', 'eu_sales', 'jp_sales', 'other_sales']
games_data['total_sales'] = games_data[sales_columns].sum(axis=1)
games_data.info()
games_data.head(30)
```

<div class="alert alert-block alert-success">
<b>Comentario de Revisor          </b> <a class="tocSkip"></a>

Bien hecho. En los casos donde hay demasiados en una columna, me parece bien dejar los nulos tal como están o reemplazarlos por algún valor identificable. En este caso, no hay data suficiente para hacer una imputación adecuada y en caso de rellenar con la media o mediana, se introduce un sesgo muy fuerte en la data, por lo que no es adecuado.



</div>

Habia un error en el tipo de dato en la columna year_of_release que se corrijio, ademas se remplazo la leyenda TBD por NaN en la columna user_score ya que esto es lo mismo que no tener calificación, despues las columnas con valores nulos se reemplazan con ceros y por ultimo se hace la suma total de las ventas de los juegos y se agregan en una columna nueva.

```python
# Primero extrae solo el año de la fecha
games_data['year'] = games_data['year_of_release'].dt.year

# Cuenta juegos por año
games_per_year = games_data['year'].value_counts().sort_index()
print(games_per_year)

# Gráfico de barras
plt.figure(figsize=(15, 6))
games_per_year.plot(kind='bar')
plt.title('Número de juegos lanzados por año')
plt.xlabel('Año')
plt.ylabel('Número de juegos')
plt.xticks(rotation=45)
plt.show()

# Estadísticas básicas
print(f"Año con más lanzamientos: {games_per_year.idxmax()} ({games_per_year.max()} juegos)")
print(f"Año con menos lanzamientos: {games_per_year.idxmin()} ({games_per_year.min()} juegos)")
print(f"Promedio de juegos por año: {games_per_year.mean():.1f}")
```

```python

# Extraer datos de géneros Action y Sports
action_scores = games_data[(games_data['genre'] == 'Action') & (games_data['user_score'] > 0)]['user_score']
sports_scores = games_data[(games_data['genre'] == 'Sports') & (games_data['user_score'] > 0)]['user_score']

print(f"Juegos de Acción con calificaciones: {len(action_scores)}")
print(f"Juegos de Deportes con calificaciones: {len(sports_scores)}")
print(f"Promedio Acción: {action_scores.mean():.2f}")
print(f"Promedio Deportes: {sports_scores.mean():.2f}")

alpha = 0.05
# Realizar la prueba t de Student
t_stat1, p_value1 = stats.ttest_ind(action_scores, sports_scores, equal_var=False)

print(f"Estadístico t: {t_stat1:.4f}")
print(f"Valor p: {p_value1:.4f}")
print(f"Nivel de significancia (α): 0.05")
if p_value1 < alpha:
    print("Rechazamos H₀")
else:
    print("No rechazamos H₀")
```
